Calculadora de IMC (Índice de Masa Corporal):

Cree una función que calcule el IMC (Índice de Masa Corporal) de una persona a partir de su peso y altura. La función debe recibir como parámetros el peso (en kg) y la altura (en cm) y devolver el valor del IMC.

Opcional: Implemente una segunda función que clasifique el IMC según los estándares de la OMS:

Bajo peso: IMC < 18.5
Peso normal: 18.5 <= IMC < 25
Sobrepeso: 25 <= IMC < 30
Obesidad: IMC >= 30

```python
def calcular_imc(peso, altura_cm):
    altura_m = altura_cm / 100  # Convertir la altura a metros
    imc = peso / (altura_m ** 2)  # Calcular el IMC
    return imc

peso = 70  # Peso en kilogramos
altura_cm = 175  # Altura en centímetros

```

```python
imc = calcular_imc(peso, altura_cm)
print(f"El IMC es: {imc:.2f}")
```

```python
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

peso = 70  # Peso en kilogramos
altura_cm = 175  # Altura en centímetros


```

```python
imc = calcular_imc(peso, altura_cm)
clasificacion = clasificar_imc(imc)

print(f"El IMC es: {imc:.2f}")
print(f"Clasificación: {clasificacion}")
```

Calculadora de área y perímetro de figuras geométricas:

Cree funciones para calcular el área y perímetro de las siguientes figuras geométricas:

Cuadrado
Triángulo (equilátero, rectángulo e isósceles)
Círculo
Rectángulo
Cada función debe recibir las dimensiones necesarias como parámetros y devolver el valor correspondiente del área o perímetro.

```python
def area_cuadrado(lado):
    return lado ** 2
def perimetro_cuadrado(lado):
    return 4 * lado
import math
```

```python
lado_cuadrado = 4
print(f"Área del cuadrado: {area_cuadrado(lado_cuadrado)}")
print(f"Perímetro del cuadrado: {perimetro_cuadrado(lado_cuadrado)}")
```

```python
def area_triangulo_equilatero(lado):
    return (math.sqrt(3) / 4) * lado ** 2
def perimetro_triangulo_equilatero(lado):
    return 3 * lado
```

```python
lado_equilatero = 4
print(f"Área del triángulo equilátero: {area_triangulo_equilatero(lado_equilatero)}")
print(f"Perímetro del triángulo equilátero: {perimetro_triangulo_equilatero(lado_equilatero)}")
```

```python
def area_triangulo_rectangulo(base, altura):
    return (base * altura) / 2
def perimetro_triangulo_rectangulo(base, altura):
    hipotenusa = math.sqrt(base ** 2 + altura ** 2)
    return base + altura + hipotenusa
```

```python
base_rectangulo = 3
altura_rectangulo = 4
print(f"Área del triángulo rectángulo: {area_triangulo_rectangulo(base_rectangulo, altura_rectangulo)}")
print(f"Perímetro del triángulo rectángulo: {perimetro_triangulo_rectangulo(base_rectangulo, altura_rectangulo)}")
```

```python
def area_triangulo_isosceles(base, lado):
    altura = math.sqrt(lado ** 2 - (base / 2) ** 2)
    return (base * altura) / 2
def perimetro_triangulo_isosceles(base, lado):
    return 2 * lado + base
```

```python
base_isosceles = 6
lado_isosceles = 5
print(f"Área del triángulo isósceles: {area_triangulo_isosceles(base_isosceles, lado_isosceles)}")
print(f"Perímetro del triángulo isósceles: {perimetro_triangulo_isosceles(base_isosceles, lado_isosceles)}")
```

```python
def area_circulo(radio):
    return math.pi * radio ** 2
def perimetro_circulo(radio):
    return 2 * math.pi * radio
```

```python
radio_circulo = 7
print(f"Área del círculo: {area_circulo(radio_circulo)}")
print(f"Perímetro del círculo: {perimetro_circulo(radio_circulo)}")
```

```python
def area_rectangulo(largo, ancho):
    return largo * ancho
def perimetro_rectangulo(largo, ancho):
    return 2 * (largo + ancho)
```

```python
largo_rectangulo = 8
ancho_rectangulo = 5
print(f"Área del rectángulo: {area_rectangulo(largo_rectangulo, ancho_rectangulo)}")
print(f"Perímetro del rectángulo: {perimetro_rectangulo(largo_rectangulo, ancho_rectangulo)}")
```
